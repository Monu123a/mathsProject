```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64
from matplotlib.collections import LineCollection


ys = 200 + np.random.randn(100)
x = np.arange(len(ys))

points = np.array([x, ys]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)


norm = plt.Normalize(ys.min(), ys.max())
lc = LineCollection(segments, cmap='plasma', norm=norm)
lc.set_array(ys)
lc.set_linewidth(2)

fig, ax = plt.subplots(figsize=(6, 4), facecolor='black')
ax.set_facecolor('black')
ax.add_collection(lc)
ax.fill_between(x, ys, 195, where=(ys > 195), facecolor='limegreen', alpha=0.3)

ax.set_xlim(x.min(), x.max())
ax.set_ylim(ys.min() - 1, ys.max() + 1)
ax.set_title("🌈 Colorful Visualization", fontsize=12, color='white')
ax.tick_params(colors='white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')

ax.grid(True, linestyle='--', alpha=0.3, color='gray')

data = io.BytesIO()
plt.savefig(data, format='png', bbox_inches='tight', facecolor=fig.get_facecolor())
image = f"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
display.display(display.Markdown(f"""![Colorful Visualization]({image})"""))
plt.close(fig)
```

```
import pandas as pd

# Load the uploaded CSV file
df = pd.read_csv("Eminem.csv")

# Preview the data
df.head()
```

```
import re
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

all_lyrics = " ".join(df["Lyric"].dropna().astype(str)).lower()
words = re.findall(r'\b\w+\b', all_lyrics)

word_counts = Counter(words)
sorted_counts = sorted(word_counts.values(), reverse=True)

ranks = np.arange(1, len(sorted_counts) + 1)
frequencies = np.array(sorted_counts)

colors = plt.cm.plasma(np.linspace(0, 1, len(frequencies)))

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')

sc = ax.scatter(ranks, frequencies, c=colors, s=10, cmap='plasma', alpha=0.9)

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_title("🎤 Zipf's Law in Eminem's Lyrics", fontsize=14, color='white')
ax.set_xlabel("Word Rank (log scale)", fontsize=12, color='white')
ax.set_ylabel("Word Frequency (log scale)", fontsize=12, color='white')

ax.tick_params(colors='white')

ax.grid(True, which="both", linestyle="--", alpha=0.4, color='gray')

cbar = plt.colorbar(sc, ax=ax, label='Rank-based Color Gradient')
cbar.ax.yaxis.set_tick_params(color='white')
plt.setp(plt.getp(cbar.ax.axes, 'yticklabels'), color='white')
cbar.set_label('Rank-based Color Gradient', color='white')

plt.tight_layout()
plt.show()
```

```
#  Step 1: Word Frequencies and Top Words
import re
from collections import Counter

all_lyrics = " ".join(df["Lyric"].dropna().astype(str)).lower()
words = re.findall(r'\b\w+\b', all_lyrics)

word_counts = Counter(words)
top_words = word_counts.most_common(20)
print("Top 20 words:\n", top_words)
```

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

sorted_counts = sorted(word_counts.values(), reverse=True)
ranks = np.arange(1, len(sorted_counts) + 1)
frequencies = np.array(sorted_counts)

log_ranks = np.log10(ranks)
log_freqs = np.log10(frequencies)
slope, intercept, r_value, _, _ = linregress(log_ranks, log_freqs)

colors = plt.cm.plasma(np.linspace(0, 1, len(frequencies)))  # vivid for dark bg

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')


sc = ax.scatter(ranks, frequencies, c=colors, s=10, cmap='plasma', label='Words', alpha=0.9)
ax.plot(ranks, 10**(intercept + slope * log_ranks), color='deepskyblue', linestyle='--',
        linewidth=2, label=f'Fit: slope = {slope:.2f}')


ax.set_xscale('log')
ax.set_yscale('log')


ax.set_title("🎧 Zipf's Law in Eminem's Lyrics with Linear Fit", fontsize=14, color='white')
ax.set_xlabel("Word Rank (log scale)", fontsize=12, color='white')
ax.set_ylabel("Word Frequency (log scale)", fontsize=12, color='white')


ax.tick_params(colors='white')
ax.grid(True, which="both", linestyle="--", alpha=0.4, color='gray')


legend = ax.legend(facecolor='black', edgecolor='white', fontsize=10)
for text in legend.get_texts():
    text.set_color('white')

plt.tight_layout()
plt.show()
```

```
from wordcloud import WordCloud
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
filtered_words = {word: count for word, count in word_counts.items() if word not in stop_words}

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='black',
    colormap='plasma'
).generate_from_frequencies(filtered_words)

plt.figure(figsize=(10, 5), facecolor='black')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("🌌 Word Cloud (Without Stopwords)", color='white', fontsize=14)
plt.tight_layout(pad=0)
plt.show()
```

```
album_groups = df.groupby("Album")

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')

for album, group in album_groups:
    lyrics = " ".join(group["Lyric"].dropna().astype(str)).lower()
    words = re.findall(r'\b\w+\b', lyrics)
    counts = sorted(Counter(words).values(), reverse=True)
    ax.plot(range(1, len(counts) + 1), counts, label=album)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title("🎶 Zipf's Law by Album", fontsize=14, color='white')
ax.set_xlabel("Rank", fontsize=12, color='white')
ax.set_ylabel("Frequency", fontsize=12, color='white')
ax.grid(True, which="both", linestyle="--", alpha=0.3, color='gray')
ax.tick_params(colors='white')

legend = ax.legend(facecolor='black', edgecolor='white')
for text in legend.get_texts():
    text.set_color('white')

description = ("This log-log plot shows **Zipf's Law** applied to Eminem's lyrics album-wise. "
               "Each line represents an album, showing how word frequency decreases with rank. "
               "A roughly linear trend confirms Zipfian behavior, though deviations may hint at stylistic changes.")
fig.text(0.5, -0.12, description, wrap=True, ha='center', fontsize=10, color='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def leading_digit(n):
    return int(str(n)[0])

first_digits = [leading_digit(freq) for freq in word_counts.values() if freq > 0]
digit_counts = Counter(first_digits)

benford_probs = {d: np.log10(1 + 1/d) for d in range(1, 10)}
total = sum(digit_counts.values())
observed = [digit_counts[d] / total for d in range(1, 10)]
expected = [benford_probs[d] for d in range(1, 10)]

bar_colors = plt.cm.Set2(np.linspace(0, 1, 9))

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')

bars = ax.bar(range(1, 10), observed, color=bar_colors, alpha=0.9, label="Observed")
ax.plot(range(1, 10), expected, 'o--', color='lime', linewidth=2, label="Benford's Law")

for i, val in enumerate(observed):
    ax.text(i + 1, val + 0.005, f"{val:.2f}", ha='center', fontsize=9, color='white')

ax.set_xticks(range(1, 10))
ax.set_xlabel("Leading Digit", fontsize=12, color='white')
ax.set_ylabel("Proportion", fontsize=12, color='white')
ax.set_title("📊 Benford's Law on Eminem's Word Frequencies", fontsize=14, color='cyan')
ax.grid(True, linestyle='--', alpha=0.3, color='gray')
ax.tick_params(colors='white')

legend = ax.legend(facecolor='black', edgecolor='white')
for text in legend.get_texts():
    text.set_color('white')

description = (
    "This chart compares the observed distribution of leading digits in Eminem's word frequencies "
    "against Benford’s Law. A close alignment suggests a natural Zipfian pattern in usage."
)
fig.text(0.5, -0.12, description, wrap=True, ha='center', fontsize=10, color='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def leading_digit(n):
    return int(str(n)[0])

lengths = df["Lyric"].dropna().astype(str).apply(lambda x: len(x.split()))
first_digits_len = [leading_digit(l) for l in lengths if l > 0]
digit_counts_len = Counter(first_digits_len)

total_len = sum(digit_counts_len.values())
observed_len = [digit_counts_len[d] / total_len for d in range(1, 10)]

benford_probs = {d: np.log10(1 + 1/d) for d in range(1, 10)}
expected = [benford_probs[d] for d in range(1, 10)]

bar_colors = plt.cm.Paired(np.linspace(0, 1, 9))

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')

bars = ax.bar(range(1, 10), observed_len, color=bar_colors, alpha=0.9, label="Observed (Song Word Counts)")
ax.plot(range(1, 10), expected, 'o--', color='lime', linewidth=2, label="Benford's Law")

for i, val in enumerate(observed_len):
    ax.text(i + 1, val + 0.005, f"{val:.2f}", ha='center', fontsize=9, color='white')

ax.set_xticks(range(1, 10))
ax.set_xlabel("Leading Digit", fontsize=12, color='white')
ax.set_ylabel("Proportion", fontsize=12, color='white')
ax.set_title("🎵 Benford's Law on Song Word Counts", fontsize=14, color='cyan')
ax.tick_params(colors='white')
ax.grid(True, linestyle='--', alpha=0.3, color='gray')

legend = ax.legend(facecolor='black', edgecolor='white')
for text in legend.get_texts():
    text.set_color('white')

description = (
    "This chart evaluates whether the number of words in each Eminem song "
    "follows Benford's Law. It analyzes the leading digit distribution of song lengths."
)
fig.text(0.5, -0.12, description, wrap=True, ha='center', fontsize=10, color='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def leading_digit(n):
    return int(str(n)[0])

all_lines = "\n".join(df["Lyric"].dropna().astype(str)).splitlines()
line_lengths = [len(line.split()) for line in all_lines if line.strip()]
line_leading_digits = [leading_digit(l) for l in line_lengths if l > 0]

digit_counts_lines = Counter(line_leading_digits)
total_lines = sum(digit_counts_lines.values())
observed_lines = [digit_counts_lines[d] / total_lines for d in range(1, 10)]

benford_probs = {d: np.log10(1 + 1/d) for d in range(1, 10)}
expected = [benford_probs[d] for d in range(1, 10)]

bar_colors = plt.cm.Set1(np.linspace(0, 1, 9))

fig, ax = plt.subplots(figsize=(10, 6), facecolor='black')
ax.set_facecolor('black')

bars = ax.bar(range(1, 10), observed_lines, color=bar_colors, alpha=0.9, label="Line Word Counts")
ax.plot(range(1, 10), expected, 'o--', color='orange', linewidth=2, label="Benford's Law")

for i, val in enumerate(observed_lines):
    ax.text(i + 1, val + 0.004, f"{val:.2f}", ha='center', fontsize=9, color='white')

ax.set_xticks(range(1, 10))
ax.set_xlabel("Leading Digit", fontsize=12, color='white')
ax.set_ylabel("Proportion", fontsize=12, color='white')
ax.set_title("📏 Benford’s Law on Lyric Line Word Counts", fontsize=14, color='aqua')
ax.tick_params(colors='white')
ax.grid(True, linestyle='--', alpha=0.3, color='gray')

legend = ax.legend(facecolor='black', edgecolor='white')
for text in legend.get_texts():
    text.set_color('white')

description = (
    "This visualization applies Benford’s Law to the number of words in each lyric line. "
    "It compares the distribution of leading digits in line word counts to Benford's expected pattern."
)
fig.text(0.5, -0.12, description, wrap=True, ha='center', fontsize=10, color='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.25)
plt.show()
```

```
def deviation_score(observed, expected):
    return np.sum(np.abs(np.array(observed) - np.array(expected)))

dev_freq = deviation_score(observed, expected)
dev_len = deviation_score(observed_len, expected)
print(f"Deviation from Benford (Word Frequencies): {dev_freq:.4f}")
print(f"Deviation from Benford (Song Lengths): {dev_len:.4f}")
```
